```python
# extract features from each photo in the directory
def extract_features(directory):
    print("Extracting features")
    model = InceptionV3()
    # restructure the model
    new_input = model.input
    hidden_layer = model.layers[-2].output 
    model = Model(new_input, hidden_layer)
    features = dict()
    print(model.summary())

    i = 0
    for name in listdir(directory):
        filename = directory + '/' + name
        #print(i, " - ", name)
        image = load_img(filename, target_size=(299, 299, 3))
        image = img_to_array(image)
        image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
        image = preprocess_input(image)
        feature = model.predict(image, verbose=0)

        image_id = name.split('.')[0]
        features[image_id] = feature
        i+=1
    print("Features extracted!")
    return features
```

```python
# saving features extracted with vgg19 from the val set
features = extract_features(PATH_DATASET)

print("Saving extracted features")
with open('features_inceptionV3_dict_val.pickle', 'wb') as f:
    pickle.dump(features, f)
print("Saved extracted features")
```
